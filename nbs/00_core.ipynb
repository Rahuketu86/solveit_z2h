How to change github remote url from http to ssh

##### Reply<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

To change your GitHub remote URL from HTTPS to SSH, you'll use the `git remote set-url` command.

First, you can check your current remote URL:

```python
!git remote -v
```

Then, to change it to SSH, you'll need to know your repository's SSH URL (which looks like `git@github.com:username/repo.git`).

##### Reply<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

Here's the command:

```python
!git remote set-url origin git@github.com:username/repo.git
```

Replace `username` with your GitHub username and `repo` with your repository name.

If your remote is named something other than `origin`, replace that too.
